```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
```

```python
# Colunas
nomes_colunas = ["Iteração", "Bubble", "Insertion", "Selection", "Busca Linear", "Busca Binária"]
# Lista dos tamanhos
tamanhos = ["100mil", "200mil", "400mil", "1600mil"]  # , "800mil"
# Dataframe
resultados = {}
```

```python
# Função para converter números formatados como string para inteiros
def converter_numero(valor):
  return int(str(valor).replace(".", "")) # remover pontos e transformar em inteiro
```

```python
# Lista das médias
medias_custo_bubble = []
medias_custo_selection = []
medias_custo_insertion = []
medias_busca_linear = []
medias_busca_binaria = []
```

```python
# Carregamento dos arquivos dos resultados
for tamanho in tamanhos:
    nome_arquivo = f"vetor{tamanho}.csv"
    df = pd.read_csv(f"vetor{tamanho}.csv", delimiter=';', skiprows=1, names=nomes_colunas, encoding="ISO-8859-1")

    # Converter strings em números
    for coluna in nomes_colunas[1:]: # pular coluna "Iteração"
        df[coluna] = df[coluna].apply(converter_numero)
    # Armazenar dataframe no dicionário
    resultados[tamanho] = df

    # Calcular médias
    medias_custo_bubble.append(df["Bubble"].mean())
    medias_custo_selection.append(df["Selection"].mean())
    medias_custo_insertion.append(df["Insertion"].mean())
    medias_busca_linear.append(df["Busca Linear"].mean())
    medias_busca_binaria.append(df["Busca Binária"].mean())

    # Exibir os dados
    print(f"Resultado do teste de {tamanho} vetores:")
    display(resultados[tamanho])
    print(f"Média Bubble: {df['Bubble'].mean()}")
    print(f"Média Selection: {df['Selection'].mean()}")
    print(f"Média Insertion: {df['Insertion'].mean()}")
    print(f"Média Busca Linear: {df['Busca Linear'].mean()}")
    print(f"Média Busca Binária: {df['Busca Binária'].mean()}")
    print("-"*100)
```

# Gráfico

```python
plt.figure(figsize=(5, 10)) # tamanho do gráfico

# Bubble Sort
plt.plot(tamanhos, medias_custo_bubble, label="Bubble", color="r")
# Selection Sort
plt.plot(tamanhos, medias_custo_selection, label="Selection", color="g", linestyle="dashed", linewidth=5)
# Insertion Sort
plt.plot(tamanhos, medias_custo_insertion, label="Insertion", color="b")

# Busca Linear
#plt.plot(tamanhos, medias_busca_linear, label="Busca Linear", color="purple", linestyle="dotted")
# Busca Binária
#plt.plot(tamanhos, medias_busca_binaria, label="Busca Binária + Ordenação", color="orange", linestyle="dashed")

# Formatar eixo Y para mostrar em bilhões (ex: 1.0B, 2.5B)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*1e-9:.0f}'))
# Definir limite do eixo Y -> cria um intervalo que melhora a visualização
plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(nbins=30))



plt.title("Algoritmos de Ordenação: Custo por Tamanho do Vetor")
plt.xlabel("Tamanho do vetor")
plt.ylabel("Média do Custo (em bilhões)")
plt.legend() # exibir legenda
plt.grid(True) # exibir grid

plt.show()
```
